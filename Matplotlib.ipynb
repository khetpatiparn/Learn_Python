##กำหนดคุณสมบัติของ Label
คุณสมบัติที่กำหนดได้มี 3 อย่าง
1. size
2. color
3. backgroundcolor

สามารถกำหนดได้ 2 แบบ

1. กำหนดแบบ dict เช่น

<code>data = {"size":20,"color":"r","backgroundcolor":"y"}<br>plt.xlabel("ชื่อ",data)</code>

2. กำหนดแบบ properties เช่น

<code>plt.xlabel("ชื่อ",size=20, color="r", backgroundcolor= "y")</code>


```python
import matplotlib.pyplot as plt
product = [15,30,10,20,60]
month = [1,2,3,4,5]
plt.plot(month, product, color="r", marker="s", linestyle=":")
data = {"size":20, "color":"b", "backgroundcolor":"y"}
plt.xlabel("Month",data)
plt.ylabel("Product",size="20", color="b", backgroundcolor="y")
plt.show()
```

##กำหนดหัวข้อด้วย Title
คุณสมบัติที่กำหนดได้มี 3 อย่าง
1. size
2. color
3. loc => center, right, left

สามารถกำหนดได้ 2 แบบ

1. กำหนดแบบ dict ใช้ได้แค่ size, color เช่น
2. กำหนดแบบ properties เช่น


```python
import matplotlib.pyplot as plt
product = [15,30,10,20,60]
month = [1,2,3,4,5]
plt.plot(month, product, color="r", marker="s", linestyle=":")
data = {"size":20, "color":"b", "backgroundcolor":"y"}
config = {"size":25, "color":"m"}
plt.xlabel("Month",data)
plt.ylabel("Product",size="20", color="b", backgroundcolor="y")
plt.title("Top Sale 2022", config, loc="left")
plt.show()
```

##Legend function
เป็นการสร้างป้ายกำหนดชื่อให้กับเส้นกราฟนั้นๆ

ใช้คำสั่ง <code>plt.legend(["product1","product2"])</code>


1. การกำหนดตำแหน่ง legend จะใช้ attribute <code>loc=""</code>
  *   <code>"upper center/left/right"</code>
  *   <code>"lower center/left/right"</code>
  *   <code>"center/left/right"</code>


2. การกำหนดขนาดข้อความใน Legend จะใช้ attribute <code>fontsize=""</code>
  * ค่าคงที่ fontsize <code>"xx-small/x-small/small/medium/large/x-large/xx-large"</code>


3. การกำหนดสีและขอบของ Legend จะใช้ attribute <code>facecolor= "สี" </code> กับ <code>edgecolor= "สี" </code>
  * facecolor : สีพื้นหลัง legend
  * edgecolor : สีเส้นขอบของ legend 


4. การกำหนด Title ของ legend มี attribute ดังนี้
  * <code>title="name"</code> : กำหนดชื่อ title
  * <code>title_fontsize= "size"</code> : กำหนดขนาดของ title
  * <code>borderpad=3</code> : ระยะระหว่างขอบกับข้อความ (default = 0.4)
  * <code>labelspacing=3</code> : ระยะระหว่างข้อความกับข้อความ (default = 0.5)

```python
import matplotlib.pyplot as plt
product1 = [15,30,10,20,60]
product2 = [10,20,30,40,50]
month = [1,2,3,4,5]
plt.plot(month, product1)
plt.plot(month, product2)
plt.xlabel("Month")
plt.ylabel("Product")
plt.title("Top Sale 2022")
plt.legend(["product1","product2"])
plt.show()
```

##Text function
ใช้คำสั่ง <code>plt.text(x,y,string,arg***)</code> โดย x,y จะเป็นพิกัดที่จะแสดงข้อความ 

เช่น <code>plt.text(0,10,"product1",size= 12)</code>

```python
import matplotlib.pyplot as plt
product1 = [15,30,10,20,60]
product2 = [10,20,30,40,50]
month = [1,2,3,4,5]
plt.plot(month, product1)
plt.plot(month, product2)
plt.xlabel("Month")
plt.ylabel("Product")
plt.title("Top Sale 2022")
plt.legend(["Mouse","Keyboard"])
plt.text(1.8,35,"Product1")
plt.text(4,15,"Product2",size= 15, color= "green")
plt.show()
```

#Pie Chart กราฟวงกลม
ใช้คำสั่ง <code>plt.pie(list/array,labels="ป้าย")</code>

มี attribute ที่บอกค่าเป็น % คือ <code>autopct="%.1f%%"</code>

```python
import matplotlib.pyplot as plt
```

```python
data = [7000,3000,800,300,200,1900]
detail = ["Rent","Grocery","Transport","Current","School fee","Savings"]
```

```python
plt.pie(data,labels=detail)
plt.show()
```

```python
plt.pie(data,labels=detail,autopct="%.1f%%")
plt.show()
```

##การปรับแต่งกราฟวงกลม
ปรับด้วย attribute
* colors : กำหนดสีให้แต่ละข้อมูล เช่น <code>colors= ["color1","color2","color3",...]</code>
* shadow : ใส่เงาให้กราฟ เช่น <code>shadow= True</code>
* explode : กำหนดระยะห่างของแต่ละข้อมูล เช่น <code>explode= [0,0,2,0.2,...]</code>
* startangle : กำหนดองศาเริ่มต้นในการแสดงผลกราฟ เช่น <code>startangle=45</code>

```python
#colors 
import matplotlib.pyplot as plt
data = [7000,3000,800,300,200,1900]
detail = ["Rent","Grocery","Transport","Current","School fee","Savings"]
colors = ["red","green","blue","magenta","cyan","yellow"]
plt.pie(data,labels=detail,colors=colors)
plt.show()
```

```python
#shadow
import matplotlib.pyplot as plt
data = [7000,3000,800,300,200,1900]
detail = ["Rent","Grocery","Transport","Current","School fee","Savings"]
colors = ["red","green","blue","magenta","cyan","yellow"]
plt.pie(data,labels=detail,colors=colors,shadow=True)
plt.show()
```

```python
#explode
import matplotlib.pyplot as plt
data = [7000,3000,800,300,200,1900]
detail = ["Rent","Grocery","Transport","Current","School fee","Savings"]
colors = ["red","green","blue","magenta","cyan","yellow"]
plt.pie(data,labels=detail,colors=colors,shadow=True,explode=[0,1,0,0.5,0.8,0])
plt.show()
```

```python
#startangle
import matplotlib.pyplot as plt
data = [7000,3000,800,300,200,1900]
detail = ["Rent","Grocery","Transport","Current","School fee","Savings"]
colors = ["red","green","blue","magenta","cyan","yellow"]
plt.pie(data,labels=detail,colors=colors,shadow=True,explode=[0,1,0,0.5,0.8,0],startangle=90)
plt.show()
```

#Scatter Plot กราฟการกระจาย
ใช้คำสั่ง <code>plt.scatter(x,y)</code>

มี attribute เพิ่มเติม เช่น

* s : กำหนดขนาดข้อมูล เช่น <code>s= [2,3,4,5,1]</code>
* c : กำหนดสีข้อมูล เช่น <code>c= ["red","pink"]</code>

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1.2,3,2.3,4.5,4.3]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
colors = ["red","blue","green","magenta","cyan"]
size = [100,1000,300,300,800]
plt.scatter(x,y,c=colors, s=size)
plt.show()
```

#Histogram
ใช้คำสั่ง <code>plt.hist()</code> ใช้กับข้อมูลที่ต่อเนื่อง 

```python
import matplotlib.pyplot as plt
import numpy as np

# สุ่มข้อมูล 5000 ชุด มี mean = 10 , sd = 2.5
data = np.random.normal(10,2.5,5000)
```

```python
# bins คือจำนวนแท่งข้อมูล
plt.hist(data, bins=500)
plt.show()
```

```python
plt.hist(data, bins=10)
plt.show()
```
