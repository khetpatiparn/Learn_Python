#สอบถามข้อมูลแบบมีเงื่อนไข
ต้องการทราบว่า
1. วันใดบ้างที่มี temp <= 20
2. วันใดบ้างที่มีแดดร้อน
3. วันใดบ้างที่ฝนตก และ temp >= 35
4. วันใดบ้างที่มีเมฆมาก หรือ temp <= 18

```python
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/learnAI with KongRukSiam/csv_file/dataupdate.xlsx","weather",index_col="Day")
```

```python
#1.วันใดบ้างที่มี temp <= 20
df[df.Temperature <= 20]
```

```python
#2.วันใดบ้างที่มีแดดร้อน
df[df.Event == "แดดร้อน"]
```

```python
#3.วันใดบ้างที่ฝนตก และ temp >= 35
df.loc[(df.Event == "ฝนตก") & (df.Temperature >= 35)]
```

```python
#4.วันใดบ้างที่มีเมฆมาก หรือ temp <= 18
df.loc[(df.Event == "เมฆมาก") | (df.Temperature <= 18)]
```

#เลือกข้อมูลด้วย isin ใช้แทน or
1. เลือกวันที่มี temp 35 องศาเท่านั้น
2. เลือกวันที่มี temp 20 หรือ 35 องศาเท่านั้น
3. เลือกวันที่มี temp 30 หรือ 35 องศาเท่านั้นและเป็นวันที่มีแดดร้อน

```python
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/learnAI with KongRukSiam/csv_file/dataupdate.xlsx","weather",index_col="Day")
```

```python
#วันใดบ้างที่มีอุณหภูมิ 20 25 30 35 องศา
#ไม่ใช้ isin
df.loc[(df.Temperature == 20) | (df.Temperature == 25) | (df.Temperature == 30) | (df.Temperature == 35)]
```

```python
#ใช้ isin
df.loc[df.Temperature.isin([20,25,30,35])]
```

```python
#1.เลือกวันที่มี temp 35 องศาเท่านั้น
df.loc[df.Temperature.isin([35])]
```

```python
#2.เลือกวันที่มี temp 20 หรือ 35 องศาเท่านั้น
df.loc[df.Temperature.isin([20,35])]
```

```python
#3.เลือกวันที่มี temp 30 หรือ 35 องศาเท่านั้นและเป็นวันที่มีแดดร้อน
df.loc[df.Temperature.isin([25,30])][df.Event == "แดดร้อน"]
```

#จัดการข้อมูลสูญหาย (Missing Values)

##ตรวจสอบข้อมูลสูญหาย
สำหรับการดูว่ามีข้อมูลส่วนใดหายไปหรือก็คือค่าว่างจะใช้ function<br>
<code>isnull()</code> มันจะคืนค่าเป็น Ture ถ้ามีค่า NaN และจะคืนเป็น False ถ้ามีข้อมูลอยู่แล้ว
<ul>
  <li><code>isnull().any()</code> : หาคอลัมน์ใดที่มีค่าว่าง</li>
  <li><code>isnull().sum()</code> : มีค่าว่างกี่จำนวนแถว</li>
</ul>

<code>notnull()</code> จะทำงานตรงข้ามกับ <code>isnull()</code> คือจะข้อมูลว่างจะ False ถ้ามีข้อมูลจะ True
<ul>
  <li><code>notnull().any()</code> : หาคอลัมน์ใดที่มีค่าข้อมูล</li>
  <li><code>notnull().sum()</code> : มีค่าที่มีข้อมูลกี่จำนวนแถว</li>
</ul>


```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/learnAI with KongRukSiam/csv_file/Employee.xlsx", index_col = "Name")
df
```

###ลบแถวบางส่วนที่มีค่าว่าง

#ลบค่าซ้ำ
เช็คค่าซ้ำด้วย function 
<code>duplicated()</code> เช่น
<code>df[df.duplicated()]</code>

และทำการลบค่าซ้ำออกไปด้วย function 
<code>drop_duplicates()</code>

```python
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/learnAI with KongRukSiam/csv_file/Employee.xlsx", index_col = "Name")
df.dropna(subset=["Job","Salary"],inplace=True)
df.dropna(axis="columns",inplace = True)
df
```

```python
# เช็คค่าซ้ำ
df[df.duplicated()]
```

```python
# ลบค่าซ้ำออกไป
df.drop_duplicates(inplace=True)
df
```
